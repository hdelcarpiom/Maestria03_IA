```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xi=[1,2,3,4,5,6,7]
yi=[0.5,2.5,2,4,3.5,6,5.5]
```

```python
xi= np.array(xi,dtype=float)
yi=np.array(yi, dtype=float)
n=len(xi)
xm = np.mean(xi)
ym= np.mean(yi)
```

```python
print ('ymedia = ',ym)
```

```python
plt.plot(xi,yi,'o',label='(xi,yi)')
plt.stem(xi,yi,bottom=ym, linefmt = ':')
plt.xlabel('xi')
plt.ylabel('yi')
plt.legend()
plt.show()
```
